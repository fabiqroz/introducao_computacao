Multiplas entradas por linha:

Entender essa sintaxe  importante, ela condensa vrios conceitos vistos at agora. Vamos interpretar essa linha de cdigo de dentro pra fora: 



1.   Primeiro comando:  `input('Digite dois valores inteiro separados por espao: ')`

> Este comando ir retornar uma *string*, afinal, o retorno de um `input()`  sempre uma *string*. 

2.   Segundo comando: `input('Digite dois valores inteiro separados por espao: ').split())`


> Este comando (funo `split()`, nativa da classe string, j conhecida por ns) ir "pegar" essa string e subdividi-la em substrings separadas por espao. Logo, se o usurio digitar algo como `1 4` como entrada. O `split()` ir; quebrar a string '1 4' em duas substrings '1'e '4'.

3. Terceiro comando: `map(int, input('Digite dois valores inteiro separados por espao: ').split())`



> este comando `(map(int, ...)`) ir **mapear** as strings '1'e '4' resultantes da quebra anterior para valores do tipo inteiro, e por fim ir retornar essas valores para as minhas variveis `a` e `b`. 


```python
a, b = map(float, input("Digite dois numeros reais: ").split(','))
```

```python
a
```

```python
b
```

```python
nome = input('Qual o seu nome? ')
idade = eval(input('Qual a sua idade? '))
peso = eval(input('Digite o seu peso (em kg): '))
altura = eval(input('Digite a sua altura (em m): '))
```

```python
imc = peso/(altura**2)
```

```python
print('{0:5} | {1:5} | {2:5}'.format('Idade', 'Peso', 'Altura'))
print('{0:5} | {1:5.2f} | {2:5.2f}'.format(idade, peso, altura))
print('\n')
print('Ola, {0}. De acordo com seus dados, seu IMC  de {1:.2f}'.format(nome, imc))
```
